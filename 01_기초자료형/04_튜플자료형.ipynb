```python
# 튜플은 어떻게 만들까?

a = ()
b = (1,) # 하나의 요소일 때 ,를 적어야함
c= (1, 2, 3)
d = 1, 2, 3 # 괄호 생략 가능
e = ("a", "b", ("ab", "cd")) # 튜플안 튜플도 가능
f = 1, # ,를 해줘야 튜플로 인식(하나의 요소일 때)
g = 1

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))

# 튜플 요소값을 삭제하려 할때
a = (1, 2, "a", "b")
# del a[0] # 튜플은 지울 수 없음
```

```python
a = 10
print(id(a))

a = 20
print(id(a))
```

```python
a = [1, 2, 3]
print(id(a))

a.append(4)
print(a)
print(id(a))
```

```python
# 튜플 인덱싱 하기
a = (1, 2, 3, 4)
print(a[0])
```

```python
# 튜플 슬라이싱 하기
a = (1, 2, 3, 4)
print(a[:2])
```

```python
# 튜플 곱하기(리스트와 같음)
a = (3, 4)
b = a*3
print(b)
```
