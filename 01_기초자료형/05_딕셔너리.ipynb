```python
# 딕셔너리 쌍 추가하기
a = {1: "a"}
a[2]="b"
print(a)
a["name"] = "ken"
print(a)

#튜플도 들어갈 수 있음
a[(1,2)] = "hobby"
print(a)

#리스트는 못들어감(변경이 가능하기 때문)
# a[[3,4]]="job"
# print(a)
```

```python
# 딕셔너리 요소 삭제하기
print(a)
del a[1]
print(a)

del a["name"]
print(a)
```

```python
# 딕셔너리에 관련된 함수
# 1. key리스트 만들기 - keys
profile = {
	"이름":"홍길동",
	"나이": 25,
	"번호": "010-1234-1234"
}

print(profile.keys())

# 2. value 리스트 만들기 - values
profile = {
	"이름":"홍길동",
	"나이": 25,
	"번호": "010-1234-1234"
}
print(profile.values())

# 3. key:value 쌍 얻기 - items
profile = {
	"이름":"홍길동",
	"나이": 25,
	"번호": "010-1234-1234"
}
print(profile.items())

# 4. key, value 쌍 모두 지우기 - clear
profile = {
	"이름":"홍길동",
	"나이": 25,
	"번호": "010-1234-1234"
}
print(profile.clear())

# 5. key로 value 얻기 - get
profile = {
	"이름":"홍길동",
	"나이": 25,
	"번호": "010-1234-1234"
}
print(profile.get("이름")) #profile["이름"]과 같은 기능


# 해당 key가 딕셔너리 안에 있는지 조사 - in
profile = {
	"이름":"홍길동",
	"나이": 25,
	"번호": "010-1234-1234"
}
print("이름" in profile)

# key로 value 얻기 - pop
profile = {
	"이름":"홍길동",
	"나이": 25,
	"번호": "010-1234-1234"
}
temp = profile.pop("이름")
print(temp)
print(profile)
```
