```python
class Calculator:		#class 이름은 파스칼형식(첫글자 대문자)으로 적어야함
	def __init__(self):		#initialize:상태값	class의 변수:property	class의 함수:method	__init__의 매개변수는 self를 해야함
		self.result = 0

	def add(self, num):
		self.result = self.result + num
		return self.result
	
	
cal1 = Calculator()		#self에 들어감
print(cal1.add(3))
print(cal1.add(4))

cal2 = Calculator()		#self.result=0이 각각 따로 갱신
print(cal2.add(5))
print(cal2.add(6))

```

```python
class Calculator:		
	def __init__(self):		
		self.result = 0

	def add(self, num):
		self.result += num
		return self.result
	
	def sub(self, num):
		self.result -= num
		return self.result
	
	def mul(self, num):
		self.result *= num
		return self.result
	
	def div(self, num):
		self.result /= num
		return self.result
	

cal1=Calculator()
cal2=Calculator()

#cal1:+10	-2		*5		/4
print(cal1.add(10))
print(cal1.sub(2))
print(cal1.mul(5))
print(cal1.div(4))

#cal2:+5	*10		-10		/5
print(cal2.add(5))
print(cal2.mul(10))
print(cal2.sub(10))
print(cal2.div(5))
	
	
```

```python
# 객체에 연산할 숫자 지정하기

class FourCal:
	def setdata(self, first, second):
		self.first = first
		self.second = second

	def add(self):
		result = self.first + self.second
		return result
	
cal3 = FourCal()
print(type(cal3))
cal3.setdata(4, 2) #first변수, second변수에 저장

print(cal3.first)
print(cal3.second)
print(cal3.add()) #괄호가 있는거는 method(함수같은거), 없는거는 property(값)
```
