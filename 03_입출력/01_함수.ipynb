```python
# 파이썬 함수의 구조
#def 함수이름(매개변수1, 매개변수2, ...):
#	수행할 문장1
#	수행할 문장2
#	...

def add(a, b):
	return a + b
```

```python
result = add(10, 20)
print(result)
```

```python
# 1. 일반적인 함수
# # 입력값 O, 반환값 O

def add(a, b):
	result = a + b
	return result

sum = add(3,4)
print(sum)

```

```python
# 2. 입력값이 없는 함수
# 입력값 X, 반환값 O

def say():
	return "hi"

a = say()
print(a)
```

```python
# 3. 반환값이 없는 함수
# 입력값 O, 반환값 X

def add(a,b):
	print(f"{a}, {b}의 합은 {a+b}입니다")

sum = add(3,4)
print(sum)
```

```python
# 4. 입력값도 없고, 반환값도 없는 함수
# 입력값 X, 반환값 X

def say():
	print("hi")

a = say()
print(a)
```
